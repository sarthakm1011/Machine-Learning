```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(1)

# initialisation kaming use it
# Batch_normalization very good to keep the side shifting of mean and covariance steady 
# Relu have surpassed tanh and sigmoid as it caters the vanishing gradient point problem as the value never becomes 0 for the delta value


#unsqueeze to increase dimensions
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())
```

```python
# Only use of for the older version
X = Variable(x, requires_grad=True)
Y = Variable(y) # actual value

torch.nn.init.kaiming_normal_(X)
```

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden1,n_hidden2, n_output):
        super(Net, self).__init__()
        self.hidden1 = torch.nn.Linear(n_feature, n_hidden1)   # hidden layer
        self.hidden2 = torch.nn.Linear(n_hidden1,n_hidden2)
        self.output = torch.nn.Linear(n_hidden2, n_output)   # output layer

    def forward(self, x):
        x = F.relu(self.hidden1(x))      # activation function for hidden layer
        x = F.relu(self.hidden2(x))
        x = self.output(x)             # linear output
        return x
    
# in init the actual class that we are overriding with this functions 
# forward is empty and we override it
# Backward method is not empty and as we write the forward pass backward is automatically created
# Lua language supports 
```

```python
net = Net(n_feature=1, n_hidden1=10,n_hidden2=5, n_output=1)     # define the network
print(net)  # net architecture
```

```python
optimizer = torch.optim.Adam(net.parameters(), lr=0.2)
loss_func = torch.nn.MSELoss()
# Gradient Descent
# SGD for each example iterate
```

```python
#plt.ion()   # something about plotting

# t:number of epochs
for t in range(250):
    prediction = net(X)     # input x and predict based on x

    loss = loss_func(prediction, Y)     # must be (1. nn output, 2. target)

    optimizer.zero_grad()   
    # clear gradients for next train if any gradients left out in the buffer
    
    loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients 

    if t % 50 == 0:
        # plot and show learning process
        plt.cla()
        plt.scatter(X.data.numpy(), Y.data.numpy())
        plt.plot(X.data.numpy(), prediction.data.numpy(), 'r-', lw=5)
        plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
        
#plt.ioff()
# plt.show()
```
