```python
import numpy as np
import matplotlib.pyplot as  plt

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)
```

```python
def f(x):
    return x * np.sin(x)
```

```python
x_plot = np.linspace(0,10,100)
print x_plot

x = np.linspace(0,10,100)
rng = np.random.RandomState(0)
rng.shuffle(x)
x = np.sort(x[:20])
y = f(x)

x_new = x[:, np.newaxis]
x_plot = x_plot[:,np.newaxis]

colors = ['teal', 'yellowgreen', 'gold','red']
line_width = 2
```

```python
plt.plot(x_plot, f(x_plot), color='cornflowerblue', linewidth=line_width, label='Ground Truth')
plt.scatter(x,y,color='navy', s=30, marker='*', label='Training Points')

non_linear_functions_y = []
for count,degree in enumerate([3,4,5,2]):
    model = make_pipeline(PolynomialFeatures(degree), Ridge())
    model.fit(x_new,y)
    y_plot = model.predict(x_plot)
    non_linear_functions_y.append(y_plot)
    plt.plot(x_plot,y_plot, color=colors[count], linewidth=line_width, label='degree %d'%degree)

plt.legend(loc='lower left')
plt.show()
```
