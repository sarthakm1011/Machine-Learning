```python
# MATPLOTLIB
# ----------------------------------

# To change the default values
params = {"ytick.color" : "w",
          "xtick.color" : "w",
          "axes.labelcolor" : "w",
          "axes.edgecolor" : "w"}
plt.rcParams.update(params)


a = np.array(range(0,100,2))
b = np.array(range(0,100,2))
print a

plt.figure(0)
plt.title("LINE",color='w')
plt.ylabel("Output",color='w')
plt.xlabel("Input Values",color='w')
plt.grid("on")
plt.xlim(0,100)
plt.ylim(0,100)
plt.plot(a,b,'r--')
plt.plot(a,2*a,'k')
plt.rcParams['axes.facecolor'] = 'w'
plt.show()
```

```python
#Creating SubPlots

x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


# First Subplot
plt.subplot(2,1,1) #set height, width and first subplot as active
plt.plot(x,y_sin)
plt.tight_layout()
plt.grid("on")
plt.tick_params(axis='x', colors='w')
plt.tick_params(axis='y', colors='w')
plt.title('Sine',color='w')

# Second Subplot
plt.subplot(2,1,2) #set height, width and first subplot as active
plt.plot(x,y_cos)
plt.tight_layout()
plt.tick_params(axis='x', colors='w')
plt.tick_params(axis='y', colors='w')
plt.title('Cosine',color='w')

plt.show()

#Learn how to create subplot grids and then using them for layout
```
